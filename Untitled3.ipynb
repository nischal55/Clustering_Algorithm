```python
import pandas as pd
def load_data():
  data = pd.read_csv('customer_segmentation_data.csv', usecols=['spending_score','purchase_frequency'])
  X = data[['spending_score', 'purchase_frequency']].values
  return X
```

```python
import numpy as np
import matplotlib.pyplot as plt
import random
from sklearn.model_selection import train_test_split
```

```python
def euclidean_distance(point1, point2):
    return np.sqrt(np.sum((point1 - point2) ** 2))

# Step 3: Initialize centroids randomly
def initialize_centroids(X, k):
    random_indices = random.sample(range(X.shape[0]), k)
    centroids = X[random_indices]
    return centroids

# Step 4: Assign clusters based on closest centroid
def assign_clusters(X, centroids):
    clusters = []
    for point in X:
        distances = [euclidean_distance(point, centroid) for centroid in centroids]
        closest_centroid = np.argmin(distances)
        clusters.append(closest_centroid)
    return np.array(clusters)

# Step 5: Recalculate centroids
def calculate_centroids(X, clusters, k):
    centroids = np.zeros((k, X.shape[1]))
    for i in range(k):
        points_in_cluster = X[clusters == i]
        if len(points_in_cluster) > 0:
            centroids[i] = np.mean(points_in_cluster, axis=0)
    return centroids

# Step 6: Implement the KMeans algorithm
def kmeans(X, k, max_iterations=100):
    centroids = initialize_centroids(X, k)
    for _ in range(max_iterations):
        clusters = assign_clusters(X, centroids)
        new_centroids = calculate_centroids(X, clusters, k)

        # Check if centroids have stopped changing
        if np.all(centroids == new_centroids):
            break
        centroids = new_centroids
    return centroids, clusters

# Step 7: Calculate the sum of squared distances for the Elbow method
def calculate_sse(X, centroids, clusters):
    sse = 0
    for i, centroid in enumerate(centroids):
        points_in_cluster = X[clusters == i]
        sse += np.sum((points_in_cluster - centroid) ** 2)
    return sse

# Step 8: Elbow method to find the optimal number of clusters
def elbow_method(X, max_k=10):
    sse_list = []
    for k in range(1, max_k+1):
        centroids, clusters = kmeans(X, k)
        sse = calculate_sse(X, centroids, clusters)
        sse_list.append(sse)

    # Plot the elbow curve
    plt.plot(range(1, max_k+1), sse_list, marker='o')
    plt.xlabel('Number of clusters (k)')
    plt.ylabel('Sum of squared distances (SSE)')
    plt.title('Elbow Method for Optimal k')
    plt.show()

    # Find the optimal k by using the "elbow" point
    optimal_k = find_elbow_point(sse_list)
    return optimal_k

# Step 9: Find the elbow point by measuring the maximum curvature
def find_elbow_point(sse_list):
    sse_diff = np.diff(sse_list)
    elbow_index = np.argmax(sse_diff[1:] - sse_diff[:-1]) + 1  # Find the point where the drop in SSE slows down
    return elbow_index + 1  # Adding 1 because np.diff reduces the index by 1

# Step 10: Split the data into training and testing sets
def split_data(X, test_size=0.2):
    X_train, X_test = train_test_split(X, test_size=test_size, random_state=42)
    return X_train, X_test

# Step 11: Evaluate KMeans on the test set by calculating the SSE
def evaluate_kmeans(X_test, centroids):
    clusters = assign_clusters(X_test, centroids)
    sse = calculate_sse(X_test, centroids, clusters)
    return sse

# Step 12: Load data, find optimal k, train on train set, and evaluate on test set
     # Replace with your CSV file path
X = load_data()

    # Split the data into training and testing sets
X_train, X_test = split_data(X, test_size=0.2)

    # Use elbow method on the training data to determine the optimal number of clusters
optimal_k = elbow_method(X_train, max_k=10)
print(f"Optimal number of clusters (k) found by elbow method: {optimal_k}")

    # Train KMeans with the optimal number of clusters on the training data
centroids, clusters_train = kmeans(X_train, optimal_k)

    # Evaluate KMeans on the test data
sse_test = evaluate_kmeans(X_test, centroids)
print(f"SSE on the test set: {sse_test}")

    # Visualize the final clustering result for the training data
plt.scatter(X_train[:, 0], X_train[:, 1], c=clusters_train, cmap='viridis', marker='o')
plt.scatter(centroids[:, 0], centroids[:, 1], s=300, c='red', label='Centroids')
plt.xlabel('Average_purchase')
plt.ylabel('Purchase_count')
plt.title(f'KMeans Clustering on Training Set with k={optimal_k}')
plt.legend()
plt.show()

    # Optionally, visualize clustering on the test set as well
clusters_test = assign_clusters(X_test, centroids)
plt.scatter(X_test[:, 0], X_test[:, 1], c=clusters_test, cmap='viridis', marker='o')
plt.scatter(centroids[:, 0], centroids[:, 1], s=300, c='red', label='Centroids')
plt.xlabel('Average_purchase')
plt.ylabel('Purchase_count')
plt.title(f'KMeans Clustering on Test Set with k={optimal_k}')
plt.legend()
plt.show()


plt.figure(figsize=(10, 6))
plt.scatter(X[:, 0], X[:, 1], c='blue', marker='o', edgecolor='k')
plt.xlabel('Average_purchase')
plt.ylabel('Purchase_count')
plt.title('Raw Data Before Clustering')
plt.grid(True)
plt.show()
```
